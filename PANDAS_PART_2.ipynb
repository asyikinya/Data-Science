```python
import pandas as pd
```

```python
#reading excel file
#read_excel()

data = pd.read_excel("https://raw.githubusercontent.com/yash240990/Python/master/sampleExcelData.xlsx")
data
```

```python
# Reading data from different data sheet
# Load entire excel workbook

excelFile = pd.ExcelFile("https://raw.githubusercontent.com/yash240990/Python/master/sampleExcelData.xlsx")
excelFile.sheet_names
```

```python
data = pd.read_excel(excelFile, "Sheet2")
data
```

```python
# Reading CSV File
# comma separeted value

data = pd.read_csv("https://raw.githubusercontent.com/yash240990/Python/master/Position_Salaries.csv")
data
```

```python
# Reading tab-deliminited file
# Reading the content of excel file with .txt format

data = pd.read_csv("https://raw.githubusercontent.com/prefuse/Prefuse/refs/heads/master/data/fisher.iris.txt")
data
```

```python
# use this methos instead, add ,sep="\t"

data = pd.read_csv("https://raw.githubusercontent.com/prefuse/Prefuse/refs/heads/master/data/fisher.iris.txt",sep="\t")
data
```

```python
# Reading a JSON

data = pd.read_json("https://raw.githubusercontent.com/yash240990/Python/master/simple.json")
data
```
