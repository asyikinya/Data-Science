```python
# DATA SCIENCE LIBRARY -- PANDAS
import numpy as np
import pandas as pd
```

```python
# numpy --> array()
arr = np.array([10,20,30,40])
print(type(arr))
print(arr)
```

```python
# Pandas Series
# Series is 1D Labelled Array

s = pd.Series([10,20,30,40])
print(type(s))
print(s)
print(s.shape)
```

```python
# in numpy array we can not store mixed data types
arr = np.array([123,'John',4500.32,True])
print(arr)
```

```python
s = pd.Series([123,'John',4500.32,True])
print(s.values)
```

```python
s = pd.Series([123,'John',4500.32,True],index=['id','name','sal','status'])
print(s)
print(s['sal'])
print(s.iloc[3])       # if you want to print the Data Value By Using An Index Number use iloc [index location]
# built-in function  get()
print(s.get('name'))

#  123   'John'    4500.32   True
#  0       1         2        3
#  id    name        sal     status

# Series is like a Dictionary + Array combo
```

```python
# Indexing Slicing And using For Loop
s = pd.Series([123,'John',4500.32,True],index=['id','name','sal','status'])
print(s)

#  123   'John'    4500.32   True
#  0       1         2        3
#  id    name        sal     status

# Indexing in Series can be done by using [] , iloc[] or loc[]
print('\nIndexing In Series')
print(s['sal'])        # use [] for labels
print(s.loc['name'])   # use loc for labels
print(s.iloc[3])       # use iloc for indexNumbers

print('\nSlicing In Series')
print(s[0:3])                 # 3 will be excluded --> stopIndex
print(s['name':'status'])    # if performing slicing using labels , the end label is included

print('\nPrinting Specific Information')
print(s[['id','name','status']])  # s[['labelname','labelname']]
print(s[['id','name','status']].values)
```

```python
# Using For Loop
for i in s:
    print(i)
```

```python
for k,v in s.items():
    print(k,v)
```
