```python
# Working with Missing Data in Pandas
# Missing Data can occur when no information is provided for one or more
# items or for a whole unit. Missing Data is a very big problem in a real-life scenarios.
# Missing Data can also refer to as NA(Not Available) values in pandas.
# In DataFrame sometimes many datasets simply arrive with missing data,
# either because it exists and was not collected or it never existed.
# For Example, Suppose different users being surveyed may choose not to share their income,
# some users may choose not to share the address in this way many datasets went missing.

# Dataset 1 : https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/ufo.csv

# Dataset 2 : https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/sample11.csv
```

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

ufo =pd.read_csv("https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/ufo.csv")
ufo.head()
```

```python
ufo.info()
```

```python
ufo.isnull().sum()
```

```python
# remove missing value
# dropna()

ufo.dropna().shape
```

```python
#fillna()
# Fill NA/NaN values using the specified method

ufo["City"].fillna(value="CityNotMentioned",inplace=True)
```

```python
ufo["City"].head(50)
```

```python
ufo["Colors Reported"].fillna(value="ColorsNotKnown",inplace=True)
ufo["Shape Reported"].fillna(value="UnknowShape",inplace=True)
```

```python
ufo.isnull().sum()
```

```python
ufo["Colors Reported"].head(50)
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/sample11.csv")
data.info()
```

```python
data.isnull().sum()
```

```python
data["Marks"].fillna(method="bfill",inplace=True) #backfill value use data from back and replace it
```

```python
data
```

```python
data["Percentage"].fillna(method="ffill",inplace=True) #forwardfill is look forward the number then Nan replace with it
```

```python
data
```

```python
# Fill Missing Values by taking up A Mean

data = pd.read_csv("https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/sample11.csv")
data
```

```python
mean_val_marks = data["Marks"].mean()
print(mean_val_marks)
```

```python
mean_val_marks = data["Marks"].mean()
data["Marks"].fillna(value=mean_val_marks,inplace=True)
data
```

```python
mean_val_marks = data["Percentage"].mean()
print(mean_val_marks)
```

```python
mean_val_per = round(data["Percentage"].mean(), 1)

data["Percentage"].fillna(mean_val_per, inplace=True)
data["Percentage"] = data["Percentage"].round(1)

data
```

```python
# Remove Duplicate Records
# Handle Duplicate Data
# Dealing With Duplicate Data

# Create a DataFrame
data = {
      'StudentName': ['Mark', 'Ali', 'Bob', 'John', 'Johny', 'Mark'],
      'Score': [45, 65, 76, 44, 39, 45]
}
df = pd.DataFrame(data)
df
```

```python
# check for duplicate records especially for thousands records
print(df.duplicated().sum())  # to know how many missing values
```

```python
df[df.duplicated()] # to show which value is duplicate
```

```python
df[df["StudentName"]=="Mark"] #to cross confirm the data that duplicate
```

```python
# dropping duplicates
df.drop_duplicates(inplace=True)
```

```python
print(df.duplicated().sum())
```

```python
# dropping duplicates
df.drop_duplicates(keep="last",inplace=True)
df
```
