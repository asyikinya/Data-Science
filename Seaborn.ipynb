```python
dataset = sns.load_dataset("iris")
dataset
```

```python
# Line Graph
# lineplot()
sns.lineplot(x="sepal_length", y="sepal_width", data=dataset)
plt.title("Lineplot In Seaborn")      # Add title graph
plt.show()
```

```python
# adding style to seaborn WHITEGRID

sns.set_style("whitegrid",{"grid.color":"lightblue","grid.linestyle":"--"})
sns.lineplot(x="sepal_length", y="sepal_width", data=dataset)
plt.title("Lineplot In Seaborn")
plt.show()
```

```python
# adding style to seaborn , DARKGRID

sns.set_style("darkgrid",{"grid.color":"lightblue","grid.linestyle":"--"})
sns.lineplot(x="sepal_length", y="sepal_width", data=dataset)
plt.title("Lineplot In Seaborn")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set_style("whitegrid",{"grid.color":"lightblue","grid.linestyle":"--"})
sns.lineplot(x="sepal_length", y="sepal_width", data=dataset)
plt.title("Lineplot In Seaborn",color="red",fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
# Heatmap
# A heatmap is a colored table where colors represent data values,
# helping to quickly identify high, low, positive, or negative values.

'''
Common places where heatmaps are used:

1. Correlation Analysis (Very common in data science)

To show relationships between variables

Example:

Study hours vs marks

Features in machine learning datasets

Helps spot strong positive or negative relationships quickly

2. Academic Performance / Education

Student scores across subjects

Attendance vs performance

Identifying strong and weak areas

3. Website & App Analytics

To see where users click or spend time

Mouse movement or scrolling behavior

Dark areas = more user activity

4. Business & Sales Analysis

Sales performance by region and time

Product demand across months

High and low revenue zones

'''

data = {
    'Hours_Studied':[2,3,4,5,6,7,8],
    'Marks_Scored':[40,50,60,70,80,90,100],
    'TV_Watched':[10,9,8,6,7,5,3]
}

df = pd.DataFrame(data)
df
```

```python
corr = df.corr()
corr
```

```python
sns.heatmap(corr,cmap="Blues",annot=True) #negative light colour, positive dark color
plt.title("Heatmap") # color 'viridis', 'coolwarm', 'Blues'
plt.show()
```
